```python
import pandas as pd
from analysis.plot import ReviewAnalysisPlots

data_dir = "../data"
plots_dir = "../plots"

topic_sentiments_df = pd.read_csv(data_dir + "/topic_sentiments.csv")
plots = ReviewAnalysisPlots(topic_sentiments_df)
```
